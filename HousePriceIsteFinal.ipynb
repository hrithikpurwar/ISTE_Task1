```python
def labelling(cats):
    df_final=df
    i=0
    for fields in cats:
        
        print(fields)
        df1=pd.get_dummies(df[fields],drop_first=True)
        
        df.drop([fields],axis=1,inplace=True)
        if i==0:
            df_final=df1.copy()
        else:
            
            df_final=pd.concat([df_final,df1],axis=1)
        i=i+1
       
        
    df_final=pd.concat([df,df_final],axis=1)
        
    return df_final
```

```python
df=labelling(category)
```

```python
from sklearn.model_selection import train_test_split
X_data=df.drop('SalePrice',axis=1)
Y_data=pd.DataFrame(df['SalePrice'])
X_train,X_test,Y_train,Y_test=train_test_split(X_data,Y_data)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
grboost=GradientBoostingRegressor(n_estimators=350,criterion='mse',max_features='sqrt',max_depth=5,min_samples_split=4)
grboost.fit(X_train,Y_train.values.ravel())
```

```python
from sklearn.model_selection import cross_val_score
scores_tr=cross_val_score(grboost,X_train,Y_train.values.ravel(),cv=5,scoring='r2')
print("for train set")
print(scores_tr)
print(np.mean(scores_tr))
```

```python
Y_pred=grboost.predict(X_test)
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(Y_test,Y_pred)
r2=r2_score(Y_test,Y_pred)
print("mean squared error: ",mse)
print("")
print("root mean squared error: ",np.sqrt(mse))
print("")
print("R sq:   ",r2)
```
